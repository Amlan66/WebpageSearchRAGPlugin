# RAG Chrome Plugin - Colab Indexer

This notebook runs on Google Colab with GPU to create and maintain FAISS index for webpage content.

## Setup Instructions:
1. Run this notebook in Google Colab with GPU runtime
2. Install dependencies
3. Start the FastAPI server with ngrok
4. Copy the ngrok URL and configure it in your local agent
5. Local agent will send webpages here for indexing
